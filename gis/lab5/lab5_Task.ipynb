# Задание для лабораторной 5


## Оценка плотности населения по городским кварталам на основе данных World Pop

В рамках этого задания вам необходимо:

1. **Создать геометрии кварталов** для одного из российских городов (подумайте, что можно использовать, чтобы сделать выделение максимально точным).
2. **Рассчитать плотность населения на основе кварталов**, используя данные World Pop.
3. **Рассчитать плотность населения на основе регулярной сетки** (квадратной или шестиугольной).
4. **Кратко описать отличия** результатов двух расчетов и особенности размещения населения внутри города.

### Требования:

- Рекомендуемый размер города — от **100 000** до **1 млн человек**.
- Важно, чтобы города в группе **не пересекались**.

Перед началом работы укажите город в [таблице](https://docs.google.com/spreadsheets/d/1GMcgG2tTWTxzXXU43UVHkkwJb3VOaTQgmm)
